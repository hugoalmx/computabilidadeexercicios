```python
def valida_placa(placa, texto):
    estado = 'q0'
    paises = ['Argentina', 'Brasil', 'Venezuela', 'Paraguai', 'Uruguai']


    if len(placa) != 7:
        return "Número da placa inválido!"


    for char in placa:
        if estado == 'q0':
            if char.isalpha():
                estado = 'q1'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q1':
            if char.isalpha():
                estado = 'q2'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q2':
            if char.isalpha():
                estado = 'q3'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q3':
            if char.isdigit():
                estado = 'q4'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q4':
            if char.isalpha():
                estado = 'q5'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q5':
            if char.isdigit():
                estado = 'q6'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q6':
            if char.isdigit():
                estado = 'q7'
            else:
                estado = 'q_rejeita'
                break


    if estado == 'q7':
        if texto in paises:
            return "Mercosul"
        else:
            return "Modelo Antigo"
    else:
        return "Placa está com caracteres fora do padrão!"

# Testes
print(valida_placa("ABC1D23", "Brasil"))  # Saída esperada: "Mercosul"
print(valida_placa("A2C1D23", "Joao Pessoa - Paraíba")) # Saída esperada: "Placa está com caracteres fora do padrão!"
print(valida_placa("A4C1D23", "Brasil")) # Saída esperada: "Placa está com caracteres fora do padrão!"
print(valida_placa("ABC1D23", "Joao Pessoa - Paraíba")) # Saída esperada: "Modelo Antigo"


```

```python
#Reconhecer placa antiga

def valida_placaantiga(placa):
    estado = 'q0'



    if len(placa) != 8:
        return "Número da placa inválido!"


    for char in placa:
        if estado == 'q0':
            if char.isalpha():
                estado = 'q1'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q1':
            if char.isalpha():
                estado = 'q2'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q2':
            if char.isalpha():
                estado = 'q3'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q3':
            if char == '-':
                estado = 'q4'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q4':
            if char.isdigit():
                estado = 'q5'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q5':
            if char.isdigit():
                estado = 'q6'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q6':
            if char.isdigit():
                estado = 'q7'
            else:
                estado = 'q_rejeita'
                break


    if estado == 'q7':
        return "Modelo Antigo"
    else:
        return "Placa está com caracteres fora do padrão!"

# Testes
print(valida_placaantiga("ABC1D23"))  # Saída esperada: Número da placa inválido!
print(valida_placaantiga("AZC-1223")) # Saída esperada: Modelo Antigo
print(valida_placaantiga("A4C-1D23")) # Saída esperada: Placa está com caracteres fora do padrão!
print(valida_placaantiga("ABC-1423")) # Saída esperada: Modelo Antigo
```

```python
def valida_cpf(cpf):
    estado = 'q0'
    cpf_mascarado = []

    # Verificar se o CPF tem exatamente 11 dígitos e se é composto apenas de números
    if len(cpf) != 11 or not cpf.isdigit():
        return "CPF inválido!"

    for i, char in enumerate(cpf):
        if estado == 'q0':
            if char.isdigit():
                cpf_mascarado.append('X')  # Mascarar o primeiro dígito
                estado = 'q1'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q1':
            if char.isdigit():
                cpf_mascarado.append('X')  # Mascarar o segundo dígito
                estado = 'q2'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q2':
            if char.isdigit():
                cpf_mascarado.append('X')  # Mascarar o terceiro dígito
                estado = 'q3'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q3':
            if char.isdigit():
                cpf_mascarado.append(char)  # Não mascarar do 4º ao 9º dígito
                estado = 'q4'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q4':
            if char.isdigit():
                cpf_mascarado.append(char)
                estado = 'q5'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q5':
            if char.isdigit():
                cpf_mascarado.append(char)
                estado = 'q6'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q6':
            if char.isdigit():
                cpf_mascarado.append(char)
                estado = 'q7'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q7':
            if char.isdigit():
                cpf_mascarado.append(char)
                estado = 'q8'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q8':
            if char.isdigit():
                cpf_mascarado.append(char)
                estado = 'q9'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q9':
            if char.isdigit():
                cpf_mascarado.append('X')  # Mascarar o penúltimo dígito
                estado = 'q10'
            else:
                estado = 'q_rejeita'
                break
        elif estado == 'q10':
            if char.isdigit():
                cpf_mascarado.append('X')  # Mascarar o último dígito
                estado = 'q11'
            else:
                estado = 'q_rejeita'
                break

    if estado == 'q11':  # Se chegou ao estado final com sucesso
        return ''.join(cpf_mascarado)
    else:
        return "CPF inválido!"

# Testes
print(valida_cpf("70468293493"))  # Saída: XXX682934XX
print(valida_cpf("12345678901"))  # Saída: XXX456789XX
print(valida_cpf("00000000000"))  # Saída: XXX000000XX
print(valida_cpf("01234567890"))  # Saída: XXX345678XX
print(valida_cpf("12345678abc"))  # Saída: CPF inválido! (não numérico)
print(valida_cpf("123456789"))    # Saída: CPF inválido! (tamanho errado)
```
